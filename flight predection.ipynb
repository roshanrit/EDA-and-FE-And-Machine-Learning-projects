```python
har["Date_of_Journey"].str.split("/").str[0]
```

```python
har["Date"]=har["Date_of_Journey"].str.split("/").str[0]
har["Month"]=har["Date_of_Journey"].str.split("/").str[1]
har["Year"]=har["Date_of_Journey"].str.split("/").str[2]
```

```python
har.head(10)
```

```python
har['Date']=har['Date'].astype(int)
har['Month']=har['Month'].astype(int)
har['Year']=har['Year'].astype(int)
```

```python
har.info()
```

```python
har.drop(["Date_of_Journey"],axis=1,inplace=True)
```

```python
har.head(10)
```

```python
har["Arrival_Time"]=har["Arrival_Time"].str.split(" ").str[0]
```

```python
har["Arrival_Time"]
```

```python
har["Arrival_Hour"]=har["Arrival_Time"].str.split(":").str[0]
har["Arrival_Min"]=har["Arrival_Time"].str.split(":").str[1]
```

```python
har.head(10)
```

```python
har["Arrival_Hour"]=har["Arrival_Hour"].astype(int)
har["Arrival_Min"]=har["Arrival_Min"].astype(int)
```

```python
har.info()
```

```python
har.drop(["Arrival_Time"],axis=1,inplace=True)
```

```python
har.head()
```

```python
har["Dep_Hour"]=har["Dep_Time"].str.split(":").str[0]
har["Dep_Min"]=har["Dep_Time"].str.split(":").str[1]
```

```python
har.head()
```

```python
har["Dep_Hour"]=har["Dep_Hour"].astype(int)
har["Dep_Min"]=har["Dep_Min"].astype(int)
```

```python
har.info()
```

```python
har.drop(["Dep_Time"],axis=1,inplace=True)
```

```python
har.head()
```

```python
har.drop(["Route"],axis=1,inplace=True)
```

```python
har.head()
```

```python
har["Total_Stops"].unique()
```

```python
har['Total_Stops']=har['Total_Stops'].map({'non-stop':1, '2 stops':2, '1 stop':3, '3 stops':4, 'nan':5, '4 stops':6})
```

```python
har.head()
```

```python
har.drop(["Duration"],axis=1,inplace=True)
```

```python
har.head()
```

```python
har["Airline"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
har["Airline"]=label.fit_transform(har["Airline"])
har["Source"]=label.fit_transform(har["Source"])
har["Destination"]=label.fit_transform(har["Destination"])
har["Additional_Info"]=label.fit_transform(har["Additional_Info"])
```
