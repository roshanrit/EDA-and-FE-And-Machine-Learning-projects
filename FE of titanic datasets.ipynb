```python
#numerical dat set and add variable in numbers
ros1=pd.read_csv("https://raw.githubusercontent.com/krishnaik06/Feature-Engineering-Live-sessions/master/titanic.csv",usecols=cat)
ros1
```

```python
cat=columns_to_use=['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked']
```

```python
from sklearn.impute import SimpleImputer
sim=SimpleImputer(strategy="most_frequent")
```

```python
ros1=sim.fit_transform(ros1)
```

```python
b=pd.DataFrame(ros1,columns=cat)
b
```

```python
c=pd.concat([a,b],axis=1)
c
```

```python
c.nunique()
```

```python
c.isnull().sum()
```

```python
#encoding for gender one hot encoding
```

```python
c["Sex"].unique()
```

```python
dum=pd.get_dummies(c["Sex"],drop_first=True)
dum
```

```python
c["Sex"]=c["Sex"].map({"male":0,"female":1})
```

```python
c
```

```python
reg=c.groupby(["Embarked"])["Age"].mean().sort_values().index
reg
```

```python
enumerate(reg,0)
```

```python
model={k:i for i,k in enumerate(reg,0)}
model
```

```python
c["Embarked"]=c["Embarked"].map(model)
```

```python
c
```

```python
#mean encoding for fare
c["Fare"].nunique()
```

```python
a=c.groupby(["Fare"])["Age"].mean().to_dict()
a
```

```python
c["Fare"]=c["Fare"].map(a)
```

```python
c
```

```python
#count/frequency encoding
c["Age"].nunique()
```

```python
val=c["Age"].value_counts().to_dict()
val
```

```python
c["Age"]=c["Age"].map(val)
```

```python
c
```

```python
q1=c["Age"].quantile(0.25)
q3=c["Age"].quantile(0.75)
```

```python
q1,q3
```

```python
iqr=q3-q1
iqr
```

```python
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
```

```python
upper_limit,lower_limit
```

```python
reg=c[(c["Age"]>lower_limit)&(c["Age"]<upper_limit)]
reg
```
